```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('marketing_campaign.csv',sep="\t")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.dropna(inplace = True)
```

```python
df["Age"] = 2021-df["Year_Birth"]
df = df.drop("Year_Birth", axis=1)
df.drop(['Z_CostContact','Z_Revenue'],axis = 1,inplace = True)

df["Spent"] = df["MntWines"]+ df["MntFruits"]+ df["MntMeatProducts"]+ df["MntFishProducts"]+ df["MntSweetProducts"]+ df["MntGoldProds"]
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
categorical_columns = ['Education','Marital_Status']

LE=LabelEncoder()
for col in categorical_columns:
    df[col]=df[[col]].apply(LE.fit_transform)
df['Dt_Customer']=df[['Dt_Customer']].apply(LE.fit_transform)
```

```python
df.info()
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler

del_cols = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
ds = df.drop(del_cols, axis=1)
scaler = StandardScaler()
scaler.fit(ds)
scaled_df = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
```

```python
# PCA
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Perform PCA for 3D plot
pca = PCA(n_components=3)
pca.fit(scaled_df)
PCA_3df = pd.DataFrame(pca.transform(scaled_df), columns=["PC1", "PC2", "PC3"])

# Extract the components for 3D plotting
x_3d = PCA_3df["PC1"]
y_3d = PCA_3df["PC2"]
z_3d = PCA_3df["PC3"]

# Plot the 3D projection
fig = plt.figure(figsize=(10, 10))
ax_3d = fig.add_subplot(121, projection="3d")
ax_3d.scatter(x_3d, y_3d, z_3d, c="blue", marker="o")
ax_3d.set_xlabel("PC1")
ax_3d.set_ylabel("PC2")
ax_3d.set_zlabel("PC3")
ax_3d.set_title("3D Projection")

# Perform PCA for 2D plot
pca = PCA(n_components=2)
pca.fit(scaled_df)
PCA_2d_df = pd.DataFrame(pca.transform(scaled_df), columns=["PC1", "PC2"])

# Extract the components for 2D plotting
x_2d = PCA_2d_df["PC1"]
y_2d = PCA_2d_df["PC2"]

# Plot the 2D projection
ax_2d = fig.add_subplot(122)
ax_2d.scatter(x_2d, y_2d, c="blue", marker="o")
ax_2d.set_xlabel("PC1")
ax_2d.set_ylabel("PC2")
ax_2d.set_title("2D Projection")

plt.tight_layout()
plt.show()
```

# Clustering

```python
# visualize best k value

from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer

def plot_clustering_evaluation(data, model, model_name):
    pca = PCA(n_components=3)
    pca_df = pca.fit_transform(data)
    
    # Elbow curve
    elbow_visualizer = KElbowVisualizer(model, k=(2, 10))
    elbow_visualizer.fit(pca_df)
    elbow_visualizer.show()
    
    print(f"Clustering evaluation for {model_name}:")
    print("Elbow value:", elbow_visualizer.elbow_value_)
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

data = PCA_3df

agglomerative_model_avg = AgglomerativeClustering(linkage='average')
agglomerative_model_ward = AgglomerativeClustering(linkage='ward')
agglomerative_model_complete = AgglomerativeClustering(linkage='complete')
agglomerative_model_single = AgglomerativeClustering(linkage='single')
kmeans_model = KMeans()
gmm_model = GaussianMixture()
```

### Aggloremative clustering

```python
plot_clustering_evaluation(data, agglomerative_model_avg, "Agglomerative average")
plot_clustering_evaluation(data, agglomerative_model_ward, "Agglomerative ward ")
plot_clustering_evaluation(data, agglomerative_model_complete, "Agglomerative complete")
plot_clustering_evaluation(data, agglomerative_model_single, "Agglomerative single")
```

```python
def plot_agglomerative_clustering(linkage_option, num_clusters):
    ac = AgglomerativeClustering(n_clusters=num_clusters, linkage=linkage_option)
    AC_df = ac.fit_predict(data)

    fig = plt.figure()

    # 2d
    ax = fig.add_subplot(121)
    ax.scatter(x_2d, y_2d, c=AC_df, cmap='tab10', marker="o")
    ax.set_xlabel('Component 1')
    ax.set_ylabel('Component 2')

    # 3d
    ax = fig.add_subplot(122, projection='3d')
    ax.scatter(x_3d, y_3d, z_3d, c=AC_df, cmap='tab10', marker="o")
    ax.set_xlabel('Component 1')
    ax.set_ylabel('Component 2')
    ax.set_zlabel('Component 3')

    plt.show()
```

```python
plot_agglomerative_clustering('average', 6)
plot_agglomerative_clustering('ward', 5)
plot_agglomerative_clustering('complete', 6)
plot_agglomerative_clustering('single', 5)
```

### KMeans

```python
plot_clustering_evaluation(data, kmeans_model, "KMeans")
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)

kmeans_labels = kmeans.labels_

fig = plt.figure()

# 2D plot
ax = fig.add_subplot(121)
ax.scatter(x_2d, y_2d, c=kmeans_labels, cmap='tab10', marker="o")
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')

# 3D plot
ax = fig.add_subplot(122, projection='3d')
ax.scatter(x_3d, y_3d, z_3d, c=kmeans_labels, cmap='tab10', marker="o")
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Component 3')

plt.show()
```

### Gaussian Mixture Model

```python
# Perform GMM clustering for different values of K
n_components = range(1, 11)  # Test K from 1 to 10
bics = []

for k in n_components:
    gmm = GaussianMixture(n_components=k)
    gmm.fit(data)
    bics.append(gmm.bic(data))

# Plot the BIC values against the number of components (clusters)
plt.plot(n_components, bics, marker='o')
plt.xlabel('Number of Components (K)')
plt.ylabel('BIC')
plt.title('BIC for Optimal K in Gaussian Mixture Model')
plt.show()
```

```python
gmm = GaussianMixture(n_components=2)
gmm.fit(data)
gmm_labels = gmm.predict(data)

fig = plt.figure()

# 2D plot
ax = fig.add_subplot(121)
ax.scatter(x_2d, y_2d, c=gmm_labels, cmap='tab10', marker="o")
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')

# 3D plot
ax = fig.add_subplot(122, projection='3d')
ax.scatter(x_3d, y_3d, z_3d, c=gmm_labels, cmap='tab10', marker="o")
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Component 3')

plt.show()
```
